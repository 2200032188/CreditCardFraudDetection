```python
x=data.drop('Class',axis=1)
y=data['Class']
#here x will be the data without Class column and y will be the Class column.
```

```python

from sklearn.model_selection import train_test_split
#from sklearn library in model_selection we import train_test_split tool.
#this train_test_split used to split the dataset into training and testing subsets.
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
log.fit(x_train,y_train)
#the model learns the relationship between x_train and y_train(target).it determines the coefficients and intercept such that predictions minimize the loss function.
#this model uses sigmoid function to map predictions and probabilities between 0 and 1.
#after this the model(log) is trained and ready to make predictions.
```

```python
y_pred1=log.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred1)
```

```python
from sklearn.metrics import precision_score,recall_score,f1_score
precision_score(y_test,y_pred1)
```

```python
recall_score(y_test,y_pred1)
```

```python
f1_score(y_test,y_pred1)
```
